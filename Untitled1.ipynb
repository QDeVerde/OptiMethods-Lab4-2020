```python
class Node:
    
    def __init__(self, path, parent_matrix):
        
        self.path = path
        self.reduced_matrix = parent_matrix.copy()
        

        
```

```python
def new_node(parent_matrix: np.ndarray, path, lvl, i, j):
    node = Node(path, parent_matrix)
    
    if lvl != 0:
        node.path.append((i, j))
        
    for k in range(parent_matrix.shape[0]):
        if lvl == 0:
            break
            
        node.reduced_matrix[i, k] = np.nan
        node.reduced_matrix[k, j] = np.nan
        
    for k in range(len(node.path)):
        node.reduced_matrix[node.path[k][1], node.path[k][0]] = np.nan
        
    node.level = lvl
    node.number = j
    return node
```

```python
def calculate_cost(initial_matrix: np.ndarray):
    local_matrix = initial_matrix.copy()
    root_weight  = 0
    
    axis0_mins = np.nanmin(local_matrix, axis=1).reshape(-1, 1)
    root_weight  += axis0_mins.sum()
    local_matrix -= axis0_mins
    
    axis1_mins = np.nanmin(local_matrix, axis=0)
    root_weight  += axis1_mins.sum()
    local_matrix -= axis1_mins
    
    return root_weight    
```

```python
def solve(matrix: np.ndarray):
    
    pq = PriorityQueue()
    path = list()
    root = new_node(matrix, path, 0, -1, 0)
    root.cost = calculate_cost(root.reduced_matrix)
    pq.add(root)
    
    while pq.size != 0:
        
        min_node = pq.pop()
        i = min_node.number
        
        if min_node.level == matrix.shape[0] - 1:
            min_node.path.add(i, 0)
            return min_node.path, min_node.cost
        
        for j in range(matrix.shape[0] - 1):
            if min_node.reduced_matrix[i, j] != np.nan:
                child = new_node(min_node.reduced_matrix, min_node.path, min_node.level + 1, i, j)
                
                child.cost = min_node.cost + min_node.reduced_matrix[i, j] + calculate_cost(child.reduced_matrix)
```
